**Dropping the inessential columns:**

```python
columns_to_drop = ['Road Network Length (in km)',
                   'Corruption Perception Index',
                   'Crime Rate (per 100,000)',
                   'Labor Force Participation Rate (%)',
                   'Tourism Revenue (in Billion USD)',
                   'Number of International Visitors (in Millions)',
                   'Number of Airports',
                   'CO2 Emissions (Million Metric Tons)',
                   'Urban Population (%)',
                   'Internet Penetration (%)',
                   'Smartphone Adoption (%)',
                   'Energy Consumption (TWh)',
                   'Doctor-to-Patient Ratio',
                   'Forest Coverage (%)']

# Get existing columns in the DataFrame
existing_columns = df.columns

# Find common columns between columns_to_drop and existing_columns
common_columns = [col for col in columns_to_drop if col in existing_columns]


df = df.drop(columns=common_columns, axis=1)  # axis=1 for columns
```

**Exploratory Data Analysis**

Let's explore the data to uncover trends and patterns in growth and development across different asian countries.

```python
#Analysis of Top 5 and Least 5 Countries by GDP with Population

# Group by 'Country' and take the maximum GDP value for each country
unique_countries_gdp = df.groupby('Country', as_index=False)[['GDP (in Trillions USD)', 'Population (in Millions)']].max() # Modified to as_index=False and include 'Population (in Millions)' in the groupby

# Top 5 and least 5 countries by GDP, including population
top_5_gdp = unique_countries_gdp.nlargest(5, 'GDP (in Trillions USD)')#[['Country', 'GDP (in Trillions USD)', 'Population (in Millions)']] # Modified to select columns after nlargest
top_5_gdp = top_5_gdp[['Country', 'GDP (in Trillions USD)', 'Population (in Millions)']] # Select the desired columns

least_5_gdp = unique_countries_gdp.nsmallest(5, 'GDP (in Trillions USD)')#[['Country', 'GDP (in Trillions USD)', 'Population (in Millions)']] # Modified to select columns after nsmallest
least_5_gdp = least_5_gdp[['Country', 'GDP (in Trillions USD)', 'Population (in Millions)']] # Select the desired columns


# Top 5 and least 5 countries by population, including GDP
top_5_population = df.nlargest(5, 'Population (in Millions)')[['Country', 'Population (in Millions)', 'GDP (in Trillions USD)']]
least_5_population = df.nsmallest(5, 'Population (in Millions)')[['Country', 'Population (in Millions)', 'GDP (in Trillions USD)']]

# Display the results
print("Top 5 Countries by GDP:")
print(top_5_gdp)

print("\nLeast 5 Countries by GDP:")
print(least_5_gdp)
```

**Visualization of Countries by GDP**

```python
#GDP v/s Country

# Group by 'Country' and take the maximum GDP value for each country
unique_countries_gdp = df.groupby('Country', as_index=False)['GDP (in Trillions USD)'].max()

# Select the top countries by GDP after grouping
top_unique_gdp = unique_countries_gdp.nlargest(6, 'GDP (in Trillions USD)')[['Country', 'GDP (in Trillions USD)']]

# Plotting the updated bar chart
plt.figure(figsize=(10, 6))
plt.bar(top_unique_gdp['Country'], top_unique_gdp['GDP (in Trillions USD)'], color='skyblue')
plt.title('Top Countries by GDP')
plt.xlabel('Country')
plt.ylabel('GDP (in Trillions USD)')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

**Insight:**

The chart shows that USA have highest GDP values (near $23 trillion in current US Dollar), suggesting relatively comparable economic sizes among these countries. Australia have the lowest GDP value.



**Correlation Analysis**

Let's examine the correlation between numeric variables to understand potential relationships.

```python
# Select only numeric columns for correlation analysis:-

important_columns = [
    'GDP (in Trillions USD)',
    'GDP per Capita (in USD)',
    'Population (in Millions)',
    'Population Growth Rate (%)',
    'Life Expectancy (Years)',
    'Unemployment Rate (%)',
]

# Filter the dataset to include only the selected columns
heatmap_df = df[important_columns]

# Calculate the correlation matrix
correlation_matrix = heatmap_df.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Heatmap of Important Columns', fontsize=16)
plt.show()
```

**Insight:**

The heatmap shows a strong negative correlation between GDP and unemployment rate (-0.65), and a positive correlation between GDP per capita and life expectancy (0.75), indicating better economic conditions improve quality of life.

Additionally, population size negatively correlates with GDP per capita (-0.68) and life expectancy (-0.60), highlighting challenges in resource distribution in larger populations.

```python
#Pie chart of agricultural land of having highest GDP rate
df_grouped = df.groupby('Country', as_index=False).max()

# Get the top N countries by GDP (you can adjust N as needed)
top_gdp_countries = df_grouped.nlargest(10, 'GDP (in Trillions USD)')  # Adjust to the top countries you want

# Extract the necessary columns for the pie chart
labels = top_gdp_countries['Country']
sizes = top_gdp_countries['Agricultural Land (%)']

# Plotting the pie chart
plt.figure(figsize=(10, 7))
plt.pie(sizes, labels=labels, pctdistance=0.8, autopct='%1.1f%%', startangle=90)
plt.title('Agricultural Land Distribution of Countries')
plt.legend(labels, title="Countries", loc="upper left")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

**Insight:**

It shows India has the largest share of agricultural land among the top GDP countries, accounting for 26.1%, followed by China with 23.5%, while Canada and Russia have the smallest shares at 3.0% and 5.6%, respectively. This indicates a significant variation in agricultural land distribution despite similar economic rankings.

```python
# Stacked bar chart for Renewable Energy Share and Agricultural Land by Country
df_grouped = df.groupby('Country').mean()

df_grouped[['Renewable Energy Share (%)', 'Agricultural Land (%)']].plot(kind='bar', stacked=True, figsize=(10, 8))
plt.title('Renewable Energy Share and Agricultural Land by Country')
plt.tight_layout()
plt.show()
```

**Insight:**

China has the highest renewable energy share at around 80%, while Australia has the lowest at around 15%. The agricultural land percentage varies across the countries, with Russia having the highest at over 60% and China having the lowest at around 20%.

```python
#Population Growth Rate by Country
plt.figure(figsize=(12, 6))

# Loop through each country to plot individual lines
countries = df['Country'].unique()
for country in countries:
    country_data = df[df['Country'] == country]
    plt.plot(country_data['Year'], country_data['Population Growth Rate (%)'], label=country)

# Title and labels
plt.title('Population Growth Rate by Country', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Population Growth Rate (%)', fontsize=12)

# Adding a legend
plt.legend(title='Country')

# Display the plot
plt.tight_layout()
plt.show()
```

**Insight:**


The graph shows that India has the highest and most stable growth rate over time, while Russia has the lowest, consistently near zero.

 Other countries like the USA, Canada, China, and Australia exhibit moderate but steady growth trends.

```python
df_grouped = df.groupby('Country', as_index=False).max()

# Get the top 5 countries by GDP
top_gdp_countries = df_grouped.nlargest(5, 'GDP (in Trillions USD)')

# Prepare the data for plotting
countries = top_gdp_countries['Country']
literacy_rate = top_gdp_countries['Literacy Rate (%)']
transport_usage = top_gdp_countries['Public Transport Usage (%)']

# Set the bar width and position of the bars
bar_width = 0.35
index = np.arange(len(countries))

# Create the bar chart
plt.figure(figsize=(12, 6))

# Bars for Literacy Rate and Public Transport Usage
plt.bar(index, literacy_rate, bar_width, label='Literacy Rate (%)', color='skyblue')
plt.bar(index + bar_width, transport_usage, bar_width, label='Public Transport Usage (%)', color='salmon')

# Title and Labels
plt.title('Literacy Rate and Public Transport Usage (Top 5 GDP Countries)', fontsize=16)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Percentage (%)', fontsize=12)
plt.xticks(index + bar_width / 2, countries, rotation=45)

# Adding a legend
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()


```

**Insight:**

The chart shows the literacy rate and public transport usage for the top GDP countries. The USA has the highest literacy rate at around 99%, while India has the highest public transport usage at around 30%.

```python
#Countries by Military Expenditure

# Clean the column names (remove any spaces or special characters)
df.columns = df.columns.str.strip()

# Group the data by 'Country' and get the maximum military expenditure for each country
df_grouped = df.groupby('Country', as_index=False).max()

# Sort the data by Military Expenditure in descending order to find the highest
df_sorted = df_grouped.nlargest(10, 'Military Expenditure (in Billion USD)')  # Adjust the column name if necessary

# Create the bar chart for the highest military expenditure
plt.figure(figsize=(12, 6))

# Bar chart of top military expenditure
plt.bar(df_sorted['Country'], df_sorted['Military Expenditure (in Billion USD)'], color='green')

# Title and labels
plt.title('Top Countries by Military Expenditure', fontsize=16)
plt.xlabel('Country')
plt.ylabel('Military Expenditure (in USD)')

# Rotate country names on x-axis if necessary
plt.xticks(rotation=90, fontsize=10)

# Display the plot
plt.tight_layout()
plt.show()
```

**Insight:**

The chart shows the top countries by military expenditure, with the USA having the highest military expenditure at over 700 billion USD, followed by China at around 250 billion USD. The remaining countries shown, including India, Russia, Australia, and Canada, all have significantly lower military expenditures compared to the top two, the USA and China.
